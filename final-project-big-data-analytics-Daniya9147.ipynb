```python
import pandas as pd
import matplotlib.pyplot as plt

final_cleaned_file_path = r"C:\Users\Дания\Desktop\END data set.csv"
cleaned_df = pd.read_csv(final_cleaned_file_path)

cleaned_df['Demand Forecast'] = pd.to_numeric(cleaned_df['Demand Forecast'], errors='coerce')
cleaned_df['Price'] = pd.to_numeric(cleaned_df['Price'], errors='coerce')
cleaned_df['Competitor Pricing'] = pd.to_numeric(cleaned_df['Competitor Pricing'], errors='coerce')
cleaned_df['Demand Forecast'].fillna(cleaned_df['Demand Forecast'].median(), inplace=True)

numeric_columns = cleaned_df.select_dtypes(include=['number'])

for column in numeric_columns.columns:
    plt.figure(figsize=(6, 3))
    cleaned_df[column].hist(bins=20, edgecolor='black')
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

```python
import matplotlib.pyplot as plt

# Group data by Region and Category to analyze sales and orders
sales_orders_analysis = cleaned_df.groupby(['Region', 'Category']).agg({
    'Units Sold': 'sum',
    'Units Ordered': 'sum'
}).reset_index()

# Plot sales and orders by Region and Category
fig, ax = plt.subplots(2, 1, figsize=(5, 6), sharex=True)

# Units Sold visualization
for region in sales_orders_analysis['Region'].unique():
    region_data = sales_orders_analysis[sales_orders_analysis['Region'] == region]
    ax[0].bar(region_data['Category'], region_data['Units Sold'], label=region)
ax[0].set_title('Units Sold by Region and Category')
ax[0].set_ylabel('Units Sold')
ax[0].legend(title="Region")
ax[0].grid(axis='y', linestyle='--', alpha=0.7)

# Units Ordered visualization
for region in sales_orders_analysis['Region'].unique():
    region_data = sales_orders_analysis[sales_orders_analysis['Region'] == region]
    ax[1].bar(region_data['Category'], region_data['Units Ordered'], label=region)
ax[1].set_title('Units Ordered by Region and Category')
ax[1].set_ylabel('Units Ordered')
ax[1].set_xlabel('Product Category')
ax[1].legend(title="Region")
ax[1].grid(axis='y', linestyle='--', alpha=0.7)

# Show the plots
plt.tight_layout()
plt.show()
```

# "Sales to Orders Ratio" (Сатылым мен Тапсырыстың Арақатынасы)
Сатылым тиімділігін бағалау: Тапсырыстардың қаншалықты орындалғанын анықтауға мүмкіндік береді.

Қойма стратегиясын жақсарту: Егер мәндер жиі 1-ден жоғары немесе 0-ге жақын болса, қойма мен тапсырыс басқару жүйесінде ақаулар болуы мүмкін.

```python
cleaned_df['Demand Forecast'] = pd.to_numeric(cleaned_df['Demand Forecast'], errors='coerce')
cleaned_df['Price'] = pd.to_numeric(cleaned_df['Price'], errors='coerce')
cleaned_df['Competitor Pricing'] = pd.to_numeric(cleaned_df['Competitor Pricing'], errors='coerce')
cleaned_df['Demand Forecast'].fillna(cleaned_df['Demand Forecast'].median(), inplace=True)

numeric_columns = cleaned_df.select_dtypes(include=['number']).columns 
cleaned_df[numeric_columns] = cleaned_df[numeric_columns].applymap(lambda x: x if x >= 0 else None)


cleaned_df['Sales to Orders Ratio'] = cleaned_df['Units Sold'] / cleaned_df['Units Ordered']

cleaned_df['Sales to Orders Ratio'].fillna(0, inplace=True)

(cleaned_df[['Units Sold', 'Units Ordered', 'Sales to Orders Ratio']].head())
```

# Ай бойынша сатылымдар графигі

```python
cleaned_df['Demand Forecast'] = pd.to_numeric(cleaned_df['Demand Forecast'], errors='coerce')
cleaned_df['Price'] = pd.to_numeric(cleaned_df['Price'], errors='coerce')
cleaned_df['Competitor Pricing'] = pd.to_numeric(cleaned_df['Competitor Pricing'], errors='coerce')
cleaned_df['Demand Forecast'].fillna(cleaned_df['Demand Forecast'].median(), inplace=True)

numeric_columns = cleaned_df.select_dtypes(include=['number']).columns 
cleaned_df[numeric_columns] = cleaned_df[numeric_columns].applymap(lambda x: x if x >= 0 else None)


# 'Date' бағанын datetime форматына түрлендіру
cleaned_df['Date'] = pd.to_datetime(cleaned_df['Date'], errors='coerce')

# Жыл-ай мәндерін шығарып алу
cleaned_df['Year-Month'] = cleaned_df['Date'].dt.to_period('M')

# Жыл-ай бойынша сатылымдарды топтау
monthly_trend = cleaned_df.groupby('Year-Month')['Units Sold'].sum().reset_index()

# График үшін 'Year-Month' мәндерін қайта жолдық форматқа түрлендіру
monthly_trend['Year-Month'] = monthly_trend['Year-Month'].astype(str)

# 2024 жылдың 1-ші айын алып тастау
monthly_trend = monthly_trend[monthly_trend['Year-Month'] != '2024-01']

# Айлық сатылым трендін визуализациялау
plt.figure(figsize=(12, 6))
plt.plot(monthly_trend['Year-Month'], monthly_trend['Units Sold'], marker='o', linestyle='-')
plt.xticks(rotation=45)
plt.title('Айлық Сатылым (Units Sold) [Без 2024-01]')
plt.xlabel('Жыл-Ай')
plt.ylabel('Сатылған өнімдер саны')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Custom функцияны анықтау
def calculate_discounted_price(row):
    return row['Price'] * (1 - row['Discount'] / 100)

# apply() қолдану
cleaned_df['Discounted Price'] = cleaned_df.apply(calculate_discounted_price, axis=1)

# Нәтижені көру
cleaned_df[['Price', 'Discount', 'Discounted Price']].head()
```
